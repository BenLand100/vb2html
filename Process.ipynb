```python
from vb2html import WebForum, SiteGen

import MySQLdb

sitename = 'SRL Archive'
description = 'Archive of the SRL Forums'
prefix='/srl/'
webroot='webroot_example'
exclude_set={30, 7, 129, 136, 6, 481, 14, 285, 91}
```

```python
def get_conn():
    return MySQLdb.connect(host='localhost', user='srl', password='srl', database='srl_archive')

conn = get_conn()
```

```python
wf = WebForum(sitename=sitename, description=description)
```

```python
wf.populate_forums(conn, exclude_set=exclude_set)
```

```python
wf.populate_threads(conn, max_threads=10)
```

```python
wf.populate_posts(get_conn, nproc=24, max_posts=10)
```

```python
import pickle

with open('srl_archive_example.pkl','wb') as out:
    pickle.dump(wf, out)
```
